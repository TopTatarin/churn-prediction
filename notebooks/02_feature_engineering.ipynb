# 02. Feature Engineering
Очистка данных и инжиниринг признаков.


```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))
```

## 1. Загрузка raw данных


```python
import pandas as pd
from src.data_loader import ensure_raw_data
from src.features import clean, engineer_features, encode_categoricals

raw_path = ensure_raw_data()
df = pd.read_csv(raw_path)
df.shape
```

## 2. Очистка (clean)
Исправление TotalCharges (blank strings у tenure == 0).


```python
df_clean = clean(df)
df_clean['TotalCharges'].isna().sum()
```

## 3. Новые признаки (engineer_features)
charges_per_month, tenure_bucket, num_services.


```python
df_fe = engineer_features(df_clean)
df_fe[['charges_per_month', 'tenure_bucket', 'num_services']].describe(include='all')
```

## 4. Кодирование категориальных признаков (OHE)


```python
df_encoded = encode_categoricals(df_fe)
df_encoded.shape
```

## 5. Сохранение processed-датасета


```python
processed_path = PROJECT_ROOT / 'data' / 'processed' / 'telco_processed.csv'
df_encoded.to_csv(processed_path, index=False)
processed_path
```
